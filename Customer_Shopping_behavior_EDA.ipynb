## 2. Handling Missing Values

```python
# 1. Calculate the category median for every row
category_medians = df.groupby('Category')['Review Rating'].transform('median')
# 2. Fill missing values using those medians
df['Review Rating'] = df['Review Rating'].fillna(category_medians)
```

```python
df.isnull().sum()
```

## 3. Redefining column names

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ' , '_')
df = df.rename(columns = {'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
# 1. Defining the exact boundary lines for the ages
# This creates ranges: (18-24], (24-34], (34-60], (60-120]
age_bins = [17, 24, 34, 60, 120]
group_labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']

# 3. Cut the data based on your custom bins
df['age_group'] = pd.cut(df['age'], bins=age_bins, labels=group_labels)
df[['age', 'age_group']].head(10)
```

```python
# Create column purchase_frequence_days
df['frequency_of_purchases'].unique()
```

```python
# Define the mapping dictionary (Text -> Days)
frequency_map = {
    'Weekly': 7,
    'Fortnightly': 14,
    'Bi-Weekly': 14,         # Usually means the same as fortnightly
    'Monthly': 30,           # Standard monthly average
    'Quarterly': 90,         # Standard 3 months
    'Every 3 Months': 90,    # Same as quarterly
    'Annually': 365          # Standard year
}

# Create the new numerical column using .map()
df['pyrchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_map)
df[['frequency_of_purchases', 'pyrchase_frequency_days']].head(10)
```
